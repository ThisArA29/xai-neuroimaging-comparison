```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

from Settings import Settings
```

```python
input_path = Settings["input_path"]
src_dir = f"{input_path}/Processed"
dataset = os.path.basename(input_path)

train_dir = os.path.join(input_path, "train")
test_dir = os.path.join(input_path, "test")

info = pd.read_csv(f"{input_path}/info.csv")
```

```python
train_df, test_df = train_test_split(
    info,
    test_size = 0.2,
    stratify = info["Group"],
    random_state = 42
)
```

```python
for _, row in train_df.iterrows():
    src = os.path.join(src_dir, row["Image_name"])
    dst = os.path.join(train_dir, row["Image_name"])
    shutil.copy(src, dst)
```

```python
for _, row in test_df.iterrows():
    src = os.path.join(src_dir, row["Image_name"])
    dst = os.path.join(test_dir, row["Image_name"])
    shutil.copy(src, dst)
```

```python
val_dir = os.path.join(input_path, "val")
os.makedirs(val_dir, exist_ok = True)
```

```python
train_files = set(os.listdir(train_dir))
train_files = [f.replace(".nii.gz","") for f in train_files]
train_info = info[info["Image_name"].isin(train_files)]
```

```python
train_df, val_df = train_test_split(
    train_info,
    test_size = 0.12,
    stratify = train_info["Group"],
    random_state = 42
)
```

```python
val_df.shape
```

```python
for _, row in val_df.iterrows():
    src = os.path.join(train_dir, f"{row['Image_name']}.nii.gz")
    dst = os.path.join(val_dir, f"{row['Image_name']}.nii.gz")
    shutil.move(src, dst)
```
